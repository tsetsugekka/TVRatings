```python
import pandas as pd
df=pd.read_csv('data/merge.csv',encoding="shift-jis")
df.head()
```

```python
df["datetime"]=pd.to_datetime(df.datetime)
```

```python
newdf=df.loc[df["datetime"].dt.hour.isin([19,20,21,22])&df["datetime"].dt.minute.isin([0,10,20,30,40,50])]
newdf
```

```python
newdf=newdf.drop(columns='Unnamed: 0')
newdf=newdf.reset_index(drop=True)
newdf
```

```python
newdf["timestamp"]=newdf.datetime.astype('int64') // 10**9
newdf
```

```python
newdf.to_csv('data/gp_each10min.csv', encoding="shift_jis")
```

```python
#以下は番組データを再整理

import pandas as pd
dfb=pd.read_csv('data/2019a.csv',encoding="shift-jis")
dfb.head()

```

```python
newdfb=dfb.loc[dfb["開始時"].isin([18,19,20,21,22])]
newdfb["datetime"]=pd.to_datetime(newdfb.datetime)
newdfb=newdfb.reset_index(drop=True)
newdfb
```

```python
import datetime

endtime=[]

for index,row in newdfb.iterrows():
 
    et=row["datetime"]+datetime.timedelta(minutes=int(row["放送分数"]))
    endtime.append(et)
    
newdfb["endtime"]=endtime
newdfb
```

```python
newdfb["datetimets"]=newdfb.datetime.astype('int64') // 10**9
newdfb["endtimets"]=newdfb.endtime.astype('int64') // 10**9


newdfb
```
