```python
import pandas as pd
import datetime
import time
start = time.time()
path="C:/Users/kari/Desktop/視聴率データ/視聴率データ/番組視聴率情報/2019/"


df=pd.read_csv(path+"20190101.pdb",header=None,sep=',',
               usecols=[1,2,4,5,6,7,8,9,13,14,27,135,136],
               names = ["曜日","局","年","月","日","開始時","開始分","放送分数",
                        "番組分類コード","新番組コード","占拠率（世帯）","メインタイトル","サブタイトル"],
               encoding="shift-jis",
              converters={"曜日":str,"局":str,"年":str,"月":str,"日":str,"開始時":str,
                          "開始分":str,"番組分類コード":str,"新番組コード":str,"メインタイトル":str,"サブタイトル":str,
                          "占拠率（世帯）":lambda x : float(x)/10,})
           
num=len(df)
df['datetime'] = None
TV={"01":"NTV","02":"TBS","03":"FTV","04":"ANB","05":"TX","06":"NHK1","07":"NHK2","08":"ETC"}
Week={"1":"月","2":"火","3":"水","4":"木","5":"金","6":"土","7":"日"}

#28h制から24h制へ
#https://teratail.com/questions/225428
for i in range(num):
    date_temp=df["年"][i]+df["月"][i]+df["日"][i]+df["開始時"][i]+df["開始分"][i]
    hour = date_temp[8:10]
    if int(hour) >= 24:
        remain = int(hour) - 23 
        date_temp_s = date_temp[:8] + "23" + date_temp[10:]
        date = datetime.datetime.strptime(date_temp_s,"%Y%m%d%H%M") + datetime.timedelta(hours=remain)
    else:
        date = datetime.datetime.strptime(date_temp, "%Y%m%d%H%M")
    
    df.loc[i,"datetime"]=date
    df.loc[i,"曜日"]=Week[df.loc[i,"曜日"]]
    df.loc[i,"局"]=TV[df.loc[i,"局"]]


end = time.time()

print('経過時間: ',end - start)

df
```
