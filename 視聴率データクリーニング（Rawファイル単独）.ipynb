```python
import pandas as pd
import datetime
import time
start = time.time()
path="C:/Users/kari/Desktop/視聴率データ/視聴率データ/毎分視聴率情報/maifen/"

df=pd.read_csv(path+"20170101.pdm",header=None,sep=',',
               usecols=[1,2,3,4,5,7,8,57,58],
               names = ["曜日","局","日付","時","分","視聴率（世帯）","視聴率（個人）","視聴数（世帯）","視聴数（個人）"],
            converters={"曜日":str,"局":str,"日付":str,"時":str,"分":str,
                           "視聴率（世帯）":lambda x : float(x)/10,
                          "視聴率（個人）":lambda x : float(x)/10,},)

num=len(df)
df['datetime'] = None
TV={"00":"HUT","01":"NTV","02":"TBS","03":"FTV","04":"ANB","05":"TX","06":"NHK1","07":"NHK2","08":"ETC"}
Week={"1":"月","2":"火","3":"水","4":"木","5":"金","6":"土","7":"日"}

#28h制から24h制へ
#https://teratail.com/questions/225428
for i in range(num):
    date_temp=df["日付"][i]+df["時"][i]+df["分"][i]
    hour = date_temp[8:10]
    if int(hour) >= 24:
        remain = int(hour) - 23 
        date_temp_s = date_temp[:8] + "23" + date_temp[10:]
        date = datetime.datetime.strptime(date_temp_s,"%Y%m%d%H%M") + datetime.timedelta(hours=remain)
    else:
        date = datetime.datetime.strptime(date_temp, "%Y%m%d%H%M")
    
    df.loc[i,"datetime"]=date
    df.loc[i,"曜日"]=Week[df.loc[i,"曜日"]]
    df.loc[i,"局"]=TV[df.loc[i,"局"]]

end = time.time()

print('経過時間: ',end - start)

df[["datetime","曜日","局","視聴率（世帯）","視聴率（個人）","視聴数（世帯）","視聴数（個人）"]]

df.to_csv('data/to_csv_out_a.csv', mode='a', header=False)
```
